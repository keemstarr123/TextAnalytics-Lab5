```python
from collections import Counter
import pandas as pd
```

```python
file_path = "Processed_Reviews.csv"
df = pd.read_csv(file_path)
```

```python
tokenized_review = df['tokenized'].dropna().apply(eval)
tokenized_review
```

```python
all_words = [word for review in tokenized_review for word in review]
unique_words = list(set(all_words))
unique_words
```

```python
word_freq = Counter(all_words)
sorted_word_freq = dict(sorted(word_freq.items(), key = lambda item: item[1], reverse=True))
sorted_word_freq
```

```python
document_vectors = []
for review in tokenized_review:
  document_vector = [1 if word in review else 0 for word in sorted_word_freq.keys()]
  document_vectors.append(document_vector)
```

```python
doc_vector_df = pd.DataFrame(document_vectors, columns=sorted_word_freq.keys())
doc_vector_df
```

```python
doc_vector_df.to_csv("document_vectors.csv", index=False)
```

```python
word_freq_df = pd.DataFrame(list(sorted_word_freq.items()), columns=["Word","Frequency"])
word_freq_df
```
